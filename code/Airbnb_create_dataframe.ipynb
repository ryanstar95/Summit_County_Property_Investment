# Airbnb Create Dataframe

Extract data from json files gathered from the Airbnb API wrapper routine and save in a dataframe.

The process:
Combine each cities files by data and save off into a combined file by date.
Next, combine each date file into one file. 
Remove any duplicates, then save file.


```python
import pandas as pd
import numpy as np

import json
```

```python
# These functions will remove data from the json file.

# remove selected dict keys from 200 block

def remove_dictionary_key_at_200_block(dict_db, temp_list):
    for list_info in temp_list:
        try:
            del dict_db["object"][list_info]
        except:
            print(f'Level 200: Item not found: {list_info} at index: {ndx}')
            
            
# remove selected dict keys from 400 block

def remove_dictionary_key_at_400_block(dict_db, temp_list):
    for list_info in temp_list:
        try:
            del dict_db["object"]["response"]["response_json"][list_info]
        except:
            print(f'Level 400: Item not found: {list_info} at index: {ndx}')
            

# remove selected dict keys from 600 block

def remove_dictionary_key_at_600_block(dict_db, temp_list):
    for ndx in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"])):
        for list_info in temp_list:
            try:
                del dict_db["object"]["response"]["response_json"]["explore_tabs"][ndx][list_info]
            except:
                print(f'Level 600: Item not found: {list_info} at index: {ndx}')
                

# remove selected dict keys from 800 block

def remove_dictionary_key_at_800_block(dict_db, temp_list):
    for ndx in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"])):
        for list_info in temp_list:
            try:
                del dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx][list_info]
            except:
                #print(f'Level 800: Item not found: {list_info} at index: {ndx}')
                continue
                
# remove selected dict keys from 1000 block

def remove_dictionary_key_at_1000_block(dict_db, temp_list):
    for ndx in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"])):
        try:
            for ndx_1 in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"])):        
                for list_info in temp_list:
                    try:
                        del dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"][ndx_1][list_info]
                    except:
                        print(f'Level 1000: Item not found: ({list_info}) at index: {ndx}')
        except:
            continue
                      
                
# remove selected dict keys from 1100 block

def remove_dictionary_key_at_1100_block(dict_db, temp_list):
    for ndx in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"])):
        try:
            for ndx_1 in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"])):        
                for list_info in temp_list:
                    try:
                        del dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"][ndx_1]["listing"][list_info]
                    except:
                        #print(f'Level 1100: Item not found: ({list_info}) at index: {ndx}')
                        continue
        except:
            continue
                
                
# remove selected dict keys from 1200 block

def remove_dictionary_key_at_1200_block(dict_db, temp_list):
    for ndx in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"])):
        try:
            for ndx_1 in range(len(dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"])):        
                for list_info in temp_list:
                    try:
                        del dict_db["object"]["response"]["response_json"]["explore_tabs"][0]["sections"][ndx]["listings"][ndx_1]["listing"]["user"][list_info]
                    except:
                        print(f'Level 1200: Item not found: ({list_info}) at index: {ndx}')
        except:
            continue                
                
                
```

```python
# list definitions

block_200_list = ['inputs', 
                  'proxy']

block_400_list = ['layout']

block_600_list = ['tab_id', 
                  'tab_name',
                  'pagination_metadata', 
                  'home_tab_metadata']

block_800_list = ['backend_search_id', 
                  'display_type', 
                  'experiments_metadata', 
                  'result_type', 
                  'search_session_id', 
                  'section_id',
                  'section_type_uid', 
                  'title', 
                  'is_paginated', 
                  'bankai_section_id', 
                  'refinements',
                  #'listings'
                  'review_items', 
                  'breadcrumbs', 
                  'section_metadata',
                  'nearby_locations', 
                  'localized_listing_count', 
                  'background_display_options', 
                  'section_component_type', 
                  'alternative_dates', 
                  'is_context_variable_validation_failed',
                  'subtitle', 
                  'see_all_info', 
                  'trip_templates'] 

block_1000_list = ['verified',
                   'verified_card']

block_1100_list = ['badges',
                   #'bathroom_label', 
                   #'bathrooms', 
                   #'bed_label', 
                   #'bedroom_label', 
                   #'bedrooms', 
                   #'beds', 
                   #'city', 
                   #'guest_label', 
                   'host_languages',
                   'host_thumbnail_url_small',
                   'host_thumbnail_url', 
                   #'id', 
                   'is_business_travel_ready', 
                   #'is_new_listing', 
                   #'is_superhost', 
                   'kicker_content', 
                   #'lat', 
                   #'lng', 
                   #'localized_city',
                   #'name', 
                   #'person_capacity', 
                   #'picture_count',
                   'picture_url', 
                   'picture_urls', 
                   'picture', 
                   'preview_amenities', 
                   'preview_encoded_png', 
                   #'property_type_id', 
                   'reviews_count', 
                   #'room_and_property_type', 
                   #'room_type_category', 
                   #'room_type', 
                   'scrim_color', 
                   'show_structured_name', 
                   #'space_type', 
                   'star_rating', 
                   #'tier_id', 
                   #'user',
                   'wide_kicker_content', 
                   #'public_address', 
                   'seo_reviews', 
                   'amenity_ids', 
                   'preview_amenity_names', 
                   'reviews', 
                   'star_rating_color', 
                   'preview_tags', 
                   'avg_rating', 
                   'map_highlight_status', 
                   'formatted_badges', 
                   'show_photo_swipe_indicator', 
                   #'min_nights', 
                   #'max_nights', 
                   'contextual_pictures', 
                   'pdp_type', 
                   'pdp_url_type', 
                   'extend_cards', 
                   'picture_ids']

block_1200_list = ['picture_url',
                   'thumbnail_url']

summit_county_cities_list = ['frisco',
                             'dillon', 
                             'breckenridge', 
                             'blue river',
                             'copper mountain',
                             'silverthorne', 
                             'summit cove',
                             'montezuma',
                             'wildernest'
                             ]


collection_date = ['20200227', 
                   '20200229', 
                   '20200301', 
                   '20200302',
                   '20200303', 
                   '20200304'                   
                  ]
```

```python
# this main driver will pull in files by city and scrub the data then comine files into one file by date

for date in collection_date:
 
    df_data = pd.DataFrame()

    for name in summit_county_cities_list:
        rawdata_file = "airbnb_rawdata_" + name.replace(" ", "") + "_" + date + ".json"

        file_name = "../rawdata/" + rawdata_file

        rawdata_json = json.load( open( file_name ) )

        #clean_up json file

    # Remove Ditionary keys
        # Level 200
        remove_dictionary_key_at_200_block(rawdata_json, block_200_list)

        # Level 400
        remove_dictionary_key_at_400_block(rawdata_json, block_400_list)

        # Level 600
        remove_dictionary_key_at_600_block(rawdata_json, block_600_list)

        # Level 800 
        remove_dictionary_key_at_800_block(rawdata_json, block_800_list)

        # Level 1000 
        remove_dictionary_key_at_1000_block(rawdata_json, block_1000_list)

        # Level 1100 
        remove_dictionary_key_at_1100_block(rawdata_json, block_1100_list)

        # Level 1200 
        remove_dictionary_key_at_1200_block(rawdata_json, block_1200_list)

    # Extract Airbnb data from json rawdata

        index_count = 0
        data = []

        for ndx_1 in range(len(rawdata_json["object"]["response"]["response_json"]["explore_tabs"])):
            try:
                for ndx_2 in range(len(rawdata_json["object"]["response"]["response_json"]["explore_tabs"][ndx_1]["sections"])):
                    try:
                        for ndx_3 in range(len(rawdata_json["object"]["response"]["response_json"]["explore_tabs"][ndx_1]["sections"][ndx_2]["listings"])):
                            try:
                                for key, value in rawdata_json["object"]["response"]["response_json"]["explore_tabs"][ndx_1]["sections"][ndx_2]["listings"][ndx_3].items():

                                    if key == 'listing':
                                        for key_1, value_1 in value.items():

                                            if key_1 == 'bathroom_label':
                                                bathroom_label = value_1
                                            if key_1 == 'bathrooms':
                                                bathrooms = value_1
                                            if key_1 == 'bed_label':
                                                bed_label = value_1
                                            if key_1 == 'bedroom_label':
                                                bedroom_label = value_1
                                            if key_1 == 'bedrooms':
                                                bedrooms = value_1

                                            if key_1 == 'beds':
                                                beds = value_1
                                            if key_1 == 'city':
                                                city = value_1
                                            if key_1 == 'guest_label':
                                                quest_label = value_1
                                            if key_1 == 'id':
                                                id_a= value_1
                                            if key_1 == 'is_new_listing':
                                                is_new_listing = value_1

                                            if key_1 == 'is_superhost':
                                                is_superhost = value_1
                                            if key_1 == 'lat':
                                                lat = value_1
                                            if key_1 == 'lng':
                                                lng = value_1
                                            if key_1 == 'localized_city':
                                                localized_city = value_1
                                            if key_1 == 'person_capacity':
                                                person_capacity = value_1

                                            if key_1 == 'property_type_id':
                                                property_type_id = value_1
                                            if key_1 == 'room_and_property_type':
                                                room_and_property_type = value_1
                                            if key_1 == 'room_type_category':
                                                room_type_category = value_1
                                            if key_1 == 'room_type':
                                                room_type = value_1
                                            if key_1 == 'space_type':
                                                space_type = value_1

                                            if key_1 == 'tier_id':
                                                tier_id = value_1
                                            if key_1 == 'public_address':
                                                public_address = value_1
                                            if key_1 == 'min_nights':
                                                min_nights = value_1
                                            if key_1 == 'max_nights':
                                                max_nights = value_1

                                            if key_1 == 'user':
                                                try:
                                                    for key_2, value_2 in value_1.items():
                                                        if key_2 == 'first_name':
                                                            user_first_name = value_2
                                                        if key_2 == 'id':
                                                            user_id = value_2
                                                        if key_2 == 'is_superhost':
                                                            user_is_superhost = value_2
                                                except:
                                                    continue

                                    if key == 'pricing_quote':
                                        for key_1, value_1 in value.items():

                                            if key_1 == 'monthly_price_factor':
                                                monthly_price_factor = value_1
                                            if key_1 == 'rate_type':
                                                rate_type = value_1
                                            if key_1 == 'weekly_price_factor':
                                                weekly_price_factor = value_1

                                            if key_1 == 'rate':
                                                try:
                                                    for key_2, value_2 in value_1.items():
                                                        if key_2 == 'amount':
                                                            rate_amount = value_2
                                                except:
                                                    continue

                                            if key_1 == 'rate_with_service_fee':
                                                try:
                                                    for key_2, value_2 in value_1.items():
                                                        if key_2 == 'amount':
                                                            rate_with_service_fee_amount = value_2
                                                except:
                                                    continue


                                data.append({'index_ndx' : index_count, 
                                             'bathroom_label': bathroom_label,
                                             'bathrooms': bathrooms,
                                             'bed_label': bed_label,
                                             'bedroom_label': bedroom_label,
                                             'bedrooms': bedrooms,

                                             'beds': beds,
                                             'city': city,
                                             'guest_label': quest_label,
                                             'id': id_a,
                                             'is_new_listing': is_new_listing,

                                             'is_superhost': is_superhost,
                                             'lat': lat,
                                             'lng': lng,
                                             'localized_city': localized_city,
                                             'person_capacity': person_capacity,

                                             'property_type_id': property_type_id,
                                             'room_and_property_type': room_and_property_type,
                                             'room_type_category': room_type_category,
                                             'room_type': room_type,
                                             'space_type': space_type,

                                             'tier_id': tier_id,
                                             'public_address': public_address,
                                             'min_nights': min_nights,
                                             'max_nights': max_nights,

                                             'monthly_price_factor': monthly_price_factor,
                                             'rate_type': rate_type,
                                             'weekly_price_factor': weekly_price_factor,
                                             'rate_amount': rate_amount,
                                             'rate_with_service_fee_amount': rate_with_service_fee_amount, 

                                             'user_first_name': user_first_name,
                                             'user_id': user_id,
                                             'user_is_superhost': user_is_superhost,

                                             'rawdata_file': rawdata_file})

                                index_count += 1

                            except:
                                continue

                    except:
                        continue

            except:
                continue

    # create a panda dataframe
        df_data_sub = pd.DataFrame(data)

    # append     
        df_data = df_data.append(df_data_sub, ignore_index = True)

#save df
    #save data
    print(f'Saving airbnb_rawdata_{date}.csv')
    df_data.to_csv('../rawdata/airbnb_rawdata_' + date + '.csv', index=False)

print(f'completed')    
```

```python
# this routine will pull in files by data and combine into one file.

#retrieve data from 
df_data_0227 = pd.read_csv('../rawdata/airbnb_rawdata_20200227.csv')
df_data_0229 = pd.read_csv('../rawdata/airbnb_rawdata_20200229.csv')
df_data_0301 = pd.read_csv('../rawdata/airbnb_rawdata_20200301.csv')
df_data_0302 = pd.read_csv('../rawdata/airbnb_rawdata_20200302.csv')
df_data_0303 = pd.read_csv('../rawdata/airbnb_rawdata_20200303.csv')
df_data_0304 = pd.read_csv('../rawdata/airbnb_rawdata_20200304.csv')

# reset the index
df_data_0227.reset_index(drop = True, inplace = True)
df_data_0229.reset_index(drop = True, inplace = True)
df_data_0301.reset_index(drop = True, inplace = True)
df_data_0302.reset_index(drop = True, inplace = True)
df_data_0303.reset_index(drop = True, inplace = True)
df_data_0304.reset_index(drop = True, inplace = True)

df_data = pd.concat([df_data_0227, df_data_0229, df_data_0301, df_data_0302, df_data_0303, df_data_0304])
```
