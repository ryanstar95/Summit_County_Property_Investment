```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from datetime import datetime

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline
```

```python
# read in data
df_data = pd.read_csv('../rawdata/summit_county_public_record_001.csv')
df_data.shape
```

```python
# Create a figure.
plt.figure(figsize=(15,7))

plt.hist(df_data['sale_price']);

plt.xlabel('Sale Price', fontsize = 20)
plt.ylabel('Number of Sales', fontsize = 20)
plt.title('Sale Price vs Number of Property Sold', fontsize = 24);
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(df_data['sale_price'])

plt.title('Box Plot of Sale Price', fontsize = 24)
plt.xlabel('Sale Price', fontsize = 20);
```

## Models

```python
X = df_data[['beds', 'baths', 'sqft']]
y = df_data['sale_price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_train)

cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()

model_score = model.score(X_test, y_test)

print(f'cv_score: {cv_score}')
print(f'model_score: {model_score}')
```
